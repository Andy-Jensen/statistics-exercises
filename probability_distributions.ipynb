```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# A bank found that the average number of cars waiting during the noon hour at a drive-up 
# window follows a Poisson distribution with a mean of 2 cars. Make a chart of this distribution and 
# answer these questions concerning the probability of cars waiting at the drive-up window.


x = range(11)
y = stats.poisson(2).pmf(x)
plt.bar(x, y, width=1, color='darkseagreen', edgecolor='black')
```

```python
# What is the probability that no cars drive up in the noon hour?
stats.poisson(2).pmf(0)
```

```python
# What is the probability that 3 or more cars come through the drive through?
stats.poisson(2).sf(2)
```

```python
# How likely is it that the drive through gets at least 1 car?
stats.poisson(2).sf(0)
```

```python
# Grades of State University graduates are normally distributed with a mean of 3.0 and a 
# standard deviation of .3. Calculate the following:
gpa = stats.norm(3, .3)
x_axis = np.arange(1, 5, 0.01)
mean = 3
sd = .3

plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

```python
# What grade point average is required to be in the top 5% of the graduating class?
gpa.isf(.05)
```

```python
# What GPA constitutes the bottom 15% of the class?
gpa.ppf(.15)
```

```python
# eccentric alumnus left scholarship money for students in the 
# third decile from the bottom of their class. Determine the range of the third decile. 
# Would a student with a 2.8 grade point average qualify for this scholarship?
gpa.ppf([.2, .3])
# yes. 2.8 would qualify!
```

```python
# If I have a GPA of 3.5, what percentile am I in?
gpa.cdf(3.5)
```

```python
# A marketing website has an average click-through rate of 2%. 
#One day they observe 4326 visitors and 97 click-throughs. 
#How likely is it that this many people or more click through?

clicks = stats.binom(4326, .02)
clicks.sf(96)
```

```python
# You are working on some statistics homework consisting of 100 questions where all of the 
# answers are a probability rounded to the hundreths place. Looking to save time, 
# you put down random probabilities as the answer to each question.

# What is the probability that at least one of your first 60 answers is correct?

guess = stats.binom(60, .01)
guess.sf(0)
```

```python
# The codeup staff tends to get upset when the student break area is not cleaned up. 
# Suppose that there's a 3% chance that any one student cleans the break area when they visit it, 
# and, on any given day, about 90% of the 3 active cohorts of 22 students visit the break area. 
```

```python
# You want to get lunch at La Panaderia, but notice that the line is usually very long at lunchtime. 
# After several weeks of careful observation, you notice that the average number of people in 
# line when your lunch break starts is normally distributed with a mean of 15 and standard deviation of 3.

# If it takes 2 minutes for each person to order, and 10 minutes from ordering to getting your food, 
# what is the likelihood that you have at least 15 minutes left to eat your food before you have to go 
# back to class? Assume you have one hour for lunch, and ignore travel time to and from La Panaderia.

people_at_lunch = stats.norm(15, 3)
x_axis = np.arange(1, 30, 0.01)
mean = 15
sd = 3

plt.plot(x_axis, norm.pdf(x_axis, mean, sd))
plt.show()
```

```python
people_at_lunch.cdf(16.5)
```
